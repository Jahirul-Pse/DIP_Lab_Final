# Task 1: Image Dimension, Frequency Distribution, and Intensity Range

```python
from google.colab import drive
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Loading the image
image_path = '/content/drive/MyDrive/Lab_Final_images/DIP_Lab_task_01.png'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Displaying the image
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()

# Image dimensions
height, width = image.shape
print(f"Image dimensions: {width} x {height}")

# Frequency distribution of intensities
flattened_image = image.flatten()
unique, counts = np.unique(flattened_image, return_counts=True)
frequency_distribution = dict(zip(unique, counts))

# Printing frequency distribution
print(f"Frequency Distribution: {frequency_distribution}")

# Intensity range
min_intensity = np.min(image)
max_intensity = np.max(image)
print(f"Intensity range: {min_intensity} - {max_intensity}")

# Plotting the histogram for intensity distribution
plt.figure(figsize=(10,6))
plt.hist(flattened_image, bins=256, range=(0, 255), color='blue', alpha=0.7)
plt.title("Histogram of Frequency Distribution")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

# Task 3: Applying median smoothing filter to image and after identify the noise

```python
# Loading the RGB image
image_rgb_path = '/content/drive/MyDrive/Lab_Final_images/DIP_Lab_task_03.png'
image_rgb = cv2.imread(image_rgb_path)

# Converting to grayscale
image_gray = cv2.cvtColor(image_rgb, cv2.COLOR_BGR2GRAY)

# Displaying the grayscale image
plt.imshow(image_gray, cmap='gray')
plt.axis('off')
plt.title("Grayscale Image with Salt-and-Pepper Noise")
plt.show()

# Checking the noise (salt-and-pepper)
# Since salt-and-pepper noise is characterized by bright and dark pixels in random locations,
# let's compute the pixel values to identify it
salt_pepper_noise = np.sum((image_gray == 255) | (image_gray == 0))
print(f"Number of salt-and-pepper noise pixels: {salt_pepper_noise}")

# Applying a median filter to remove salt-and-pepper noise
image_smooth = cv2.medianBlur(image_gray, 3)  # You can adjust the kernel size (3x3)

# Displaying the smoothed image
plt.imshow(image_smooth, cmap='gray')
plt.axis('off')
plt.title("Image After Median Filtering (Noise Removed)")
plt.show()
```

# Task 4: Reducing brightness of the image using OpenCV scaling technique

```python
# Loading the image (DIP_Lab_task_04.jpg)
image_path = '/content/drive/MyDrive/Lab_Final_images/DIP_Lab_task_04.jpg'
image = cv2.imread(image_path)

# Converting the image to rgb
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Displaying the original RGB image
plt.imshow(image_rgb)
plt.axis('off')
plt.title("Original Image (RGB)")
plt.show()

# Reducing brightness by scaling the pixel values by 0.8 using OpenCV
brightness_reduced_image = cv2.convertScaleAbs(image_rgb, alpha=0.8, beta=0)

# Displaying the brightness-reduced RGB image
plt.imshow(brightness_reduced_image)
plt.axis('off')
plt.title("Brightness Reduced Image (OpenCV Scaling)")
plt.show()
```

# Task 5: Applied Laplacian of Gaussian (LoG) edge detection for image segmentation

```python
# Loading the image (DIP_Lab_task_05.jpg)
image_path = '/content/drive/MyDrive/Lab_Final_images/DIP_Lab_task_05.jpg'
image = cv2.imread(image_path)

# Converting the image to grayscale
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Displaying the original RGB image
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Original Image (RGB)")
plt.show()

# Applying Gaussian filter
blurred_image = cv2.GaussianBlur(image_gray, (5, 5), 0)

# Applying Laplacian edge detection
laplacian = cv2.Laplacian(blurred_image, cv2.CV_64F)

# Displaying the edges
plt.imshow(laplacian, cmap='gray')
plt.axis('off')
plt.title("Laplacian of Gaussian (LoG) Edge Detection")
plt.show()
```
