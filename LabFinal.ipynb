```python
# Task 1: Image Dimension, Frequency Distribution, and Intensity Range
from google.colab import drive
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Loading the image
image_path = '/content/drive/MyDrive/Lab_Final_images/DIP_Lab_task_01.png'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Displaying the image
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()

# Image dimensions
height, width = image.shape
print(f"Image dimensions: {width} x {height}")

# Frequency distribution of intensities
flattened_image = image.flatten()
unique, counts = np.unique(flattened_image, return_counts=True)
frequency_distribution = dict(zip(unique, counts))

# Printing frequency distribution
print(f"Frequency Distribution: {frequency_distribution}")

# Intensity range
min_intensity = np.min(image)
max_intensity = np.max(image)
print(f"Intensity range: {min_intensity} - {max_intensity}")

# Plotting the histogram for intensity distribution
plt.figure(figsize=(10,6))
plt.hist(flattened_image, bins=256, range=(0, 255), color='blue', alpha=0.7)
plt.title("Histogram of Frequency Distribution")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
